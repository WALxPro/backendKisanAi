```python
from tensorflow.keras.models import load_model

model = load_model("D:\code\KisanAI\kisanai-backend\kisanai-backend\cnn_model\cnn_model.h5")
```

```python
import os

train_dir = "D:\\code\\KisanAI\\dataset\\train"

class_names = sorted(os.listdir(train_dir))
print(class_names)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

img_path = "D:\\code\\KisanAI\\dataset\\test\\Corn___Northern_Leaf_Blight\\image (4).JPG"  # 👈 change this

# Load image
img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img)

# Normalize (IMPORTANT)
img_array = img_array / 255.0

# Add batch dimension
img_array = np.expand_dims(img_array, axis=0)

# Predict
prediction = model.predict(img_array)

predicted_index = np.argmax(prediction)
predicted_class = class_names[predicted_index]
confidence = np.max(prediction)

print("Prediction:", predicted_class)
print("Confidence:", confidence * 100, "%")
```
